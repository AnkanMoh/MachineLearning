```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    df = df[~((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)))]
    return df

alpha = remove_outliers(alpha, 'price_usd')

# Verify the outliers are removed
plt.figure(figsize=(10, 6))
sns.boxplot(data=alpha)
plt.title('Box plot after removing outliers')
plt.show()
```

```python
brand_counts = alpha['brand_name'].value_counts()

top_5_brands = brand_counts.head(5)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_5_brands.index, y=top_5_brands.values)
plt.title('Top 5 Brands Analysis')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(['Top 5 Brands'], loc='upper right')
plt.show()
```

```python
gender_counts = alpha['product_gender_target'].value_counts()

# Set up the figure and axes for the plots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Bar chart
axes[0].bar(gender_counts.index, gender_counts.values, color=['pink', 'blue'])  # Assuming male is represented by blue and female by pink
axes[0].set_title('Purchases by Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Number of Purchases')
axes[0].set_xticks(range(len(gender_counts.index)))
axes[0].set_xticklabels(gender_counts.index, rotation=0)

# Pie chart
colors = ['blue' if gender == 'Men' else 'pink' for gender in gender_counts.index]
axes[1].pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
axes[1].set_title('Purchases by Gender')
axes[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the plots
plt.tight_layout()
plt.show()

# Print the majority gender target
majority_gender = gender_counts.idxmax()
print(f"The majority gender target is: {majority_gender}")
```

```python
category_counts = alpha['product_category'].value_counts()
```

```python
color_dict = {
    'Men Clothing': 'blue',
    'Women Clothing': 'orange',
    'Women Accessories': 'green',
    'Women Shoes': 'yellow',
    'Men Shoes': 'red',
    'Men Accessories': 'purple',
}

# Split the categories into Men and Women
men_categories = category_counts[category_counts.index.str.startswith('Men')]
women_categories = category_counts[category_counts.index.str.startswith('Women')]
```

```python
#men's category
plt.figure(figsize=(10, 6))
men_categories.plot(kind='bar', color=[color_dict.get(category, 'pink') for category in men_categories.index])
plt.title('Purchases by Men Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8, 8))
plt.pie(men_categories.values, labels=men_categories.index, autopct='%1.1f%%', startangle=140, colors=[color_dict.get(category, 'pink') for category in men_categories.index])
plt.title('Purchases by Men Product Categories')
plt.axis('equal')
plt.show()

plt.figure(figsize=(10, 6))
men_categories.plot(kind='barh', color=[color_dict.get(category, 'pink') for category in men_categories.index])
plt.title('Purchases by Men Product Categories')
plt.xlabel('Number of Purchases')
plt.ylabel('Product Category')
plt.show()
```

```python
#Women's Category
plt.figure(figsize=(10, 6))
women_categories.plot(kind='bar', color=[color_dict.get(category, 'pink') for category in women_categories.index])
plt.title('Purchases by Women Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8, 8))
plt.pie(women_categories.values, labels=women_categories.index, autopct='%1.1f%%', startangle=140, colors=[color_dict.get(category, 'pink') for category in women_categories.index])
plt.title('Purchases by Women Product Categories')
plt.axis('equal')  
plt.show()

plt.figure(figsize=(10, 6))
women_categories.plot(kind='barh', color=[color_dict.get(category, 'pink') for category in women_categories.index])
plt.title('Purchases by Women Product Categories')
plt.xlabel('Number of Purchases')
plt.ylabel('Product Category')
plt.show()
```

```python
most_demanded_men = men_categories.idxmax()
most_demanded_women = women_categories.idxmax()

# Print conclusions
print(f"The most demanded product category for Men is: {most_demanded_men}")
print(f"The most demanded product category for Women is: {most_demanded_women}")

# Determine the overall most demanded category
overall_most_demanded = category_counts.idxmax()

print(f"The overall most demanded product category is: {overall_most_demanded}")
```

```python
alpha['count']=alpha['product_season'].value_counts()
```

```python
sold_and_available = alpha[(alpha['sold'] == 1) & (alpha['available'] == 1)]
sold_but_not_available = alpha[(alpha['sold'] == 1) & (alpha['available'] == 0)]

# Count the products in each category
count_sold_and_available = len(sold_and_available)
count_sold_but_not_available = len(sold_but_not_available)

print("Count of products when they are sold and available:", count_sold_and_available)
print("Count of products when they are sold but not available:", count_sold_but_not_available)
```

```python
alpha['gender_category'] = alpha['product_category'].str.split().str[0]

# Calculate profit for each product
alpha['profit'] = alpha['price_usd'] - alpha['seller_earning']

# Group by product category and calculate total profit
total_profit_by_category = alpha.groupby('product_category')['profit'].sum()

# Create a pivot table to separate categories into men and women
pivot_table = alpha.pivot_table(index='product_category', columns='gender_category', values='profit', aggfunc='sum', fill_value=0)

# Determine which category gives maximum profit in each gender category
most_profitable_men_category = pivot_table['Men'].idxmax()
most_profitable_women_category = pivot_table['Women'].idxmax()

# Plotting total profit for each category with a colorful palette
plt.figure(figsize=(10, 6))
sns.barplot(x=total_profit_by_category.index, y=total_profit_by_category.values, palette='husl')
plt.title('Total Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Profit (USD)')
plt.xticks(rotation=45)
plt.show()

# Print the table
print("Profit Table by Category:")
print(pivot_table)

# Print the most profitable categories in each gender category
print(f"\nThe most profitable category for men is: {most_profitable_men_category}")
print(f"The most profitable category for women is: {most_profitable_women_category}")
max_profit_category = total_profit_by_category.idxmax()
print(f"The product category that gives maximum profit is: {max_profit_category}")
```

```python
plt.figure(figsize=(10, 6))
top_materials = alpha['product_material'].value_counts().nlargest(3)
colors = sns.color_palette("husl", 3)  # Using HUSL color palette
top_materials.plot(kind='bar', color=colors)
plt.title('Top 3 Most Demanded Product Materials')
plt.xlabel('Product Material')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plotting top 3 most demanded product colors
plt.figure(figsize=(10, 6))
top_colors = alpha['product_color'].value_counts().nlargest(3)
colors = ['black', 'blue', 'grey']  # Assigning colors to the top 3 colors
top_colors.plot(kind='bar', color=colors)
plt.title('Top 3 Most Demanded Product Colors')
plt.xlabel('Product Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Determine the most demanded product material and color
most_demanded_material = top_materials.idxmax()
most_demanded_color = top_colors.idxmax()

print(f"The most demanded product material is: {most_demanded_material}")
print(f"The most demanded product color is: {most_demanded_color}")
```
