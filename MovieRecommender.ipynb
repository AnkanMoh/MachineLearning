```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('dataset.csv')
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies.info()
```

```python
movies['Tags']= movies['genre'] + ' '+  movies['overview']
```

```python
movies.head()
```

```python
movies_new_df=movies[['id','title','genre','Tags','overview']]
```

```python
movies_new_df=movies_new_df.drop(columns=['genre','overview'])
```

```python
movies_new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vec = cv.fit_transform(movies_new_df['Tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

```python
movies_new_df[movies_new_df['title']== 'The Godfather']
```

```python
dist = sorted(list(enumerate(sim[0])),reverse= True,key=lambda vec:vec[1])
```

```python
dist
```

```python
for i in dist[0:5]:
  print(movies_new_df.iloc[i[0]].title)
```

```python
def recommend(movies):
  index= movies_new_df[movies_new_df['title']==movies].index[0]
  distance = sorted(list(enumerate(sim[index])),reverse= True,key=lambda vec:vec[1])
  for i in distance[0:5]:
    print(movies_new_df.iloc[i[0]].title)
```

```python
recommend("Heropanti")
```
