```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
cars=pd.read_csv('/content/used_cars.csv')
```

```python
cars.head()
```

```python
cars['milage'] = cars['milage'].str.replace(' mi.', '').str.replace(',', '').astype(float)
cars['price'] = cars['price'].str.replace('$', '').str.replace(',', '').astype(float)
cars['clean_title'] = cars['clean_title'].fillna('Unknown')
cars=cars.drop(['engine'],axis=1)
cars.head()
```

```python
cat_col=['brand','model','fuel_type','transmission','ext_col','int_col','accident','clean_title']
cars.encoded=pd.get_dummies(cars,columns=cat_col,drop_first=True)
```

```python
if 'id' not in cars.encoded.columns:
    cars.encoded['id'] = range(len(cars.encoded))
```

```python
X=cars.encoded.drop(['price','id'],axis=1)
y=cars.encoded['price']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=2)
```

```python
test_ids = cars.encoded.loc[X_test.index, 'id']
```

```python
mod=LinearRegression()
mod.fit(X_train,y_train)
```

```python
y_pred = mod.predict(X_test)
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)
print(r2)
print(f'Predicted price is {rmse}')
```

```python
results = pd.DataFrame({
    'id': test_ids,
    'price': rmse
})

print(results)
```

```python
results.to_csv('submission_file.csv', index=False)
```
